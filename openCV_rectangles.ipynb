# Import all libs

```python
from pathlib import Path

import cv2
from cv2 import imread
import matplotlib.pyplot as plt
import numpy as np
import openvino as ov
```

## load img and apply filters

```python
img = imread('img.png')
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(imgray, (25,25), 1)
plt.imshow(imgray)
```

## get contours from img

```python
edges = cv2.Canny(imgray, 50, 150, apertureSize=3)
contours, hierarchy = cv2.findContours(edges, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

### apply the rectangles to the img using the contours

```python
for c in contours: 
    x, y, w, h = cv2.boundingRect(c)
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(img, "Rectangle", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
# Display the result
plt.imshow(img)
```
